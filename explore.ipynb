```python
import pandas as pd 
import email
import email.policy
from email.message import EmailMessage
from email.parser import BytesParser, Parser
```

```python
raw_emails = pd.read_csv("emails.csv")
```

```python
raw_emails
```

```python
email_seg = raw_emails.sample(n=10000)
headers = Parser(policy=email.policy.default).parsestr(email_seg.iloc[0]['message'])
for tag in headers:
    print(f'{tag} : {headers[tag]}')
```

```python
def get_body(row):
    message = ''
    lines = row['message'].split('\n')
    for line in lines:
        if ":" not in line:
            message += line.strip()
    return message


def get_headers(row):
  
    headers = Parser(policy=email.policy.default).parsestr(row['message'])
    #I dont know why this has to be out of order, but it does
    return pd.Series([headers['from'], headers['subject'], headers['x-bcc'],headers['date'], headers['x-cc'], headers['to']])

cooked_emails = email_seg.apply(get_headers, axis=1)

cooked_emails['body'] = email_seg.apply(get_body, axis=1)


```

```python
cooked_emails.rename(columns={0:"from", 1:"subject", 2:"bcc",3:"date",4:"cc",5:"to"}, inplace=True)
cooked_emails
```

```python
def remove_emails_from_body(row):
    
    body = str(row['body'])
    array_of_words = body.split(" ")
   
    for i in range(len(array_of_words)):
        if '@' in array_of_words[i]: 
            array_of_words[i] = ""
        if ',' in array_of_words[i]:
            array_of_words[i] = array_of_words[i].replace(",", " ")
        array_of_words[i] = "".join(filter(str.isalnum, array_of_words[i]))
    

    return " ".join(array_of_words)
```

```python
cooked_emails['body'] = cooked_emails.apply(remove_emails_from_body, axis=1)
cooked_emails.iloc[1]['body']
cooked_emails
```

```python
cooked_emails.to_csv("parsed_emails.csv")
```
